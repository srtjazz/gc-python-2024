```python
import pandas as pd
```

```python
filmes = pd.read_csv("movies.csv")
filmes.columns = ['filme_id', 'titulo', 'generos']
filmes.head()
```

```python
filmes.shape
```

```python
#binary the genre
generos = filmes.generos.str.get_dummies()
generos.head()
```

```python
#concatena os dados binarizados com o restante do dataset
dados_dos_filmes = pd.concat([filmes, generos], axis=1)
dados_dos_filmes.head()
```

```python
dados_dos_filmes.describe()
```

```python
#Normaliza os dados StandardScaler or MinMax
#MinMax não resolve aqui pois já possui uma distribuição próxima a Gaussiana
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
scaler = StandardScaler()
generos_escalados = scaler.fit_transform(generos)
generos_escalados[:3,:]
```

```python
scaler_min_max = preprocessing.MinMaxScaler()
data_scaled = scaler_min_max.fit_transform(generos)
data_scaled[:3,:]
```

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=3)
modelo.fit(generos_escalados)
print(f'Grupos {modelo.labels_}')
```

```python
type(modelo.labels_)
```

```python
dados_dos_filmes['Cluster'] = modelo.labels_
dados_dos_filmes.head()
```

```python
dados_dos_filmes.groupby("Cluster").size()
```

```python
print(generos.columns)
print(modelo.cluster_centers_)
```

```python
grupos = pd.DataFrame(modelo.cluster_centers_,columns=generos.columns)
grupos
```

```python
#subplot: make separate plot for each column
grupos.transpose().plot.bar(subplots=True,figsize=(25, 25),sharex=False)
```

```python
#the first 10 records from cluster 0
grupo = 0
filtro = modelo.labels_ == grupo
dados_dos_filmes[filtro].sample(10)
```

```python
#t-distributed Stochastic Neighbor Embedding
from sklearn.manifold import TSNE
import time
```

```python
#reduz a dimensionalidade para uma melhor visualização
time_start = time.time()
tsne = TSNE()
visualizacao = tsne.fit_transform(generos_escalados)
print ('t-SNE done! Time elapsed: {} seconds',format(time.time()-time_start))
visualizacao
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize': (13, 13)})
sns.scatterplot(x=visualizacao[:, 0],
               y=visualizacao[:, 1],
               hue=modelo.labels_,
               palette=sns.color_palette('Set1', 3))
```

```python
modelo = KMeans(n_clusters=20)

modelo.fit(generos_escalados)

grupos = pd.DataFrame(modelo.cluster_centers_,
            columns=generos.columns)

grupos.transpose().plot.bar(subplots=True,
               figsize=(25, 50),
               sharex=False,
               rot=0)
```

```python
grupo = 2
filtro = modelo.labels_ == grupo
dados_dos_filmes[filtro].sample(10)
```

```python
def kmeans(numero_de_clusters, generos):
  modelo = KMeans(n_clusters=numero_de_clusters)
  modelo.fit(generos)
  return [numero_de_clusters, modelo.inertia_/100]
```

```python
kmeans(20, generos_escalados)
```

```python
kmeans(3, generos_escalados)
```

```python
resultado=[]
for numero_de_grupos in range (1,41):
    resultado.append(kmeans(numero_de_grupos, generos_escalados))
resultado
```

```python
resultado = pd.DataFrame(resultado, columns=['grupos', 'inertia'], index=list(range(1, 41)))
resultado
```

```python
resultado.inertia.plot(x=resultado.grupos, y=resultado.inertia,figsize=(20, 5),xticks=resultado.grupos)
#xlabel="Grupos", ylabel="SSE"
```
