✦ Use 3 dataframes criados na aula anterior, mostre 
apenas o filme/série/anime com maior pontuação. 

```python
import pandas as pd

series_df = pd.read_json('../aula_03-04/series.json')
movies_df = pd.read_json('../aula_03-04/movies.json')
animes_df = pd.read_json('../aula_03-04/animes.json')

df_merged = pd.concat([series_df, movies_df, animes_df])

df_merged['Notas'].max()
```

✦ Utilize uma função para identificar o valor médio

```python
media_df = df_merged['Notas'].mean()
media_df
```

✦ Filtre o dataframe pegando apenas as linhas com valores maiores que essa média.

```python
df_merged[df_merged['Notas'] > media_df]
```
