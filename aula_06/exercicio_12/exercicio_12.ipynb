# Exercício 12 - Inner Join
#### - Objetivo: Unir os DataFrames de livros e empréstimos usando o método inner join e mostrar apenas os livros que foram emprestados.</br>
#### 1.Carregue os arquivos livros.csv e emprestimos.csv em DataFrames separados.

```python
import pandas as pd

books_df = pd.read_csv("../livros.csv")
books_df

loans_df = pd.read_csv("../emprestimos.csv")
loans_df
```

#### 2. Realize um inner join entre as tabelas, unindo-as pela coluna id_livro.

```python
df_merged = pd.merge(books_df, loans_df, on='id_livro', how='inner')
```

#### 3. Exiba o resultado, mostrando as colunas: titulo, autor, nome_leitor, data_emprestimo e data_devolucao.

```python
df_merged[['titulo', 'autor', 'nome_leitor', 'data_emprestimo', 'data_devolucao']]
```
