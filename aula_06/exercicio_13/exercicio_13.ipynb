# Exerc√≠cio 13 - Left Join
#### - Objetivo: Realizar um left join para exibir todos os livros, independentemente de terem sido emprestados.

#### 1. Carregue novamente os arquivos.

```python
import pandas as pd

books_df = pd.read_csv("../livros.csv")
books_df

loans_df = pd.read_csv("../emprestimos.csv")
loans_df
```

#### 2. Realize um left join entre as tabelas livros e emprestimos pela coluna id_livro.

```python
df_merged = pd.merge(books_df, loans_df, on='id_livro', how='left')
df_merged
```

#### 3. Exiba os livros que n√£o foram emprestados (os que t√™m valores NaN nas colunas relacionadas aos empr√©stimos).

```python
df_merged[df_merged['data_emprestimo'].isnull()]
```
