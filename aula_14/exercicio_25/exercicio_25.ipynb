###  Utilize o dataset zoo_2.csv e execute o algoritmo utilizando o método holdout, reservando apenas 20% dos dados para teste.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeClassifier

zoo_df = pd.read_csv('../exercicio_25/zoo_2.csv')
zoo_df
```

```python
X = zoo_df[['hair', 'feathers', 'eggs', 'milk', 'airborne', 'aquatic', 'predator',
       'toothed', 'backbone', 'breathes', 'venomous', 'fins', 'legs', 'tail',
       'domestic', 'catsize']]
y = zoo_df['type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
confusion_train = confusion_matrix(y_train, y_train_pred)
confusion_test = confusion_matrix(y_test, y_test_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_train)
disp.plot(cmap=plt.cm.GnBu)
plt.show()
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_test)
disp.plot(cmap=plt.cm.GnBu)
plt.show()
```

```python
print("Treino: \n", classification_report(y_train, y_train_pred))
```

```python
print("Teste: \n", classification_report(y_test, y_test_pred))
```
