## Você foi designado para consolidar e preparar dois arquivos de dados de Pokémon para análise: O objetivo é unificar os dados em um único DataFrame e realizar uma série de operações de limpeza e filtragem para garantir a integridade do conjunto de dados.


### Tarefas a serem realizadas: </br>
### - Unificar os arquivos: Combine os dados de pokedex_1.csv e pokedex_2.json em um único DataFrame, garantindo que as informações estejam corretamente alinhadas.

```python
import pandas as pd

df_pokedex = pd.read_csv('pokedex_(Update_04.21).csv')
df_pokedex
```

### - Remoção de duplicatas: Verifique a existência de registros duplicados no DataFrame consolidado e elimine-os para garantir a unicidade dos dados.

```python
#Exclui das duplicatas criadas a partir do id do pokemon
df_drop = df_pokedex.drop_duplicates(subset='pokedex_number').reset_index()
df_drop
```

### - Remover colunas irrelevantes: Analise as colunas presentes e exclua aquelas que não são relevantes para a análise


```python
df_column_drop = df_drop.drop(columns=['index','Unnamed: 0', 'german_name', 'japanese_name'])
df_column_drop
```

#### - Tratar valores ausentes: Identifique valores nulos ou ausentes e aplique as ações necessárias para corrigi-los (preenchimento ou remoção)

```python
df_column_drop.isna().sum().values
```

```python
df_column_drop.fillna('Não há valor aqui')
```

#### - Filtrar Pokémon proibidos: Exclua do conjunto de dados todos os Pokémon que pertencem às seguintes categorias:
#### • Lendários e Sub-lendários
#### • Míticos
#### • Pokémon com Mega Evoluções


```python
df_filtered = df_column_drop[
  (~df_column_drop['status'].isin(['Sub Legendary', 'Legendary', 'Mythical']) 
  & ~df_column_drop['name'].str.contains('Mega', na=False))
]

df_filtered
```
